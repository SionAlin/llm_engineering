and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports
import os
import requests
import json
from typing import List
from dotenv import load_dotenv
from bs4 import BeautifulSoup
from IPython.display import Markdown, display, update_display
from openai import OpenAI
import ollama
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up environment
load_dotenv()
openai = OpenAI()
```

```python
# here is the question; type over this to ask something new
question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
```

```python
system_prompt = "You are a tutor that can answer evry question clearly"
user_prompt = "Please answer me this question: " + question
```

```python
message = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt},
]
```

```python
response = openai.chat.completions.create(model=MODEL_GPT, messages=message)
result = response.choices[0].message.content
display(Markdown(result))
```

```python
# Get Llama 3.2 to answer
response = ollama.chat(model=MODEL_LLAMA, messages=message)
result = response["message"]["content"]
display(Markdown(result))
```
